### Code to remove the background of 3 views



```python
import requests
list_names = ["background_remove1.png","background_remove2.png","background_remove13.png"]

for i in range(len(list1)):
  response = requests.post('https://api.remove.bg/v1.0/removebg',files={'image_file': open(list1[i], 'rb')},
                           data={'size': 'auto'},headers={'X-Api-Key': '4Xbaftf1XrJqNeyywEK4eJ9g'},)
  if response.status_code == requests.codes.ok:
    with open(list_names[i], 'wb') as out:
      out.write(response.content)
  else:
    print("Error:", response.status_code, response.text)
```
